```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def make_image(lm):
    arr = []
    for y in range(0, 256):
        arr.append([lm(x, y) for x in range(0, 256)])
    return np.array(arr)
```

```python
X = make_image(lambda x,y : x ^ y)
C = make_image(lambda x,y : math.cos(math.pi * x / 255.0 - math.pi * y / 255.0))
```

```python
def show_image(img):
    plt.figure(figsize=(4, 4))
    plt.imshow(img)
    plt.show()

show_image(C)
show_image(X)
show_image(make_image(lambda x, y: 510 - abs(x - y)))
```
